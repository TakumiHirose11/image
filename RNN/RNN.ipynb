## Load Data

```python
import os
import json

file_path = "./projectnextnlp-chat-dialogue-corpus/json/rest1046/"
print(file_path)
file_dir = os.listdir(file_path)
print(file_path)

utterance_txt = []
system_txt = []
label = []

for file in file_dir:
    "open json"
    r = open(file_path + file, "r", encoding="utf-8")
    json_data = json.load(r)

    for turn in json_data["turns"]:
        turn_index = turn["turn-index"] #index of turn
        speaker = turn["speaker"] #speaker key(S:system, U:human)
        utterance = turn["utterance"] #text

        #except first utterance (system utteramce)
        if turn_index != 0:
            if speaker == "U":
                u_text = utterance
            
            else:
                a = ""
                sys = turn["utterance"]
                t = turn["annotations"][0]
                a = t["breakdown"]
                if a == 'O':
                    val = 0
                elif a == "T":
                    val  = 1
                else:
                    val = 2 # if a=="x"
                utterance_txt.append(u_text)
                system_txt.append(sys)
                label.append(str(val))
```

# Dicomposition into morphemes

```python
from janome.tokenizer import Tokenizer
import re

def parse(utterance_txt):
    t = Tokenizer()
    separation_tmp = []
    for row in utterance_txt:
        tokens = t.tokenize(row)

        separation_tmp.append(
            [token.surface for token in list(tokens) if(
                not re.match("記号", token.part_of_speech)
                and (not re.match("助詞", token.part_of_speech))
                and (not re.match("助動詞", token.part_of_speech))
            )]
        )

        while separation_tmp.count("") > 0:
            separation_tmp.remove("")
    
    return separation_tmp

"""形態素解析"""
df["utterance_token"] = parse(df["utterance_txt"])
df["system_token"] = parse(df["system_txt"])

df
```

## Make dictionary {word: number of appearance}

```python
from collections import Counter
import itertools


def makedictionary(data):
    return Counter(itertools.chain(* data))
```

## Function sorting by order of appearance and numbering

```python
def update_word_dictionary(worddic):
    word_list = []
    word_dic = {}
    for w in worddic.most_common():
        word_list.append(w[0])
    
    for i, word in enumerate(word_list, start=1):
        word_dic.update({word: i})
    
    return word_dic
```

```python
def backOfWords(word_dic, token):
    return [[ word_dic[word] for word in sp] for sp in token]
```
